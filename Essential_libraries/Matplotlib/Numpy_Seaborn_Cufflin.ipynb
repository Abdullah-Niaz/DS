```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x ** 2
```

```python
plt.plot(x,y)
plt.title("My title")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
```

```python
plt.subplot(2,2,1)
plt.plot(x,y)
plt.subplot(2,2,4)
plt.plot(x,y)
plt.subplot(2,2,3)
plt.plot(y,x)
plt.subplot(2,2,2)
plt.plot(x**2,x**2)
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
# [left, right, width, height]
axes1.plot(x,y)

axes2 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2.plot(x,y)
```

```python
 plt.scatter(x,y)
```

```python
plt.boxplot(x)
```

```python
 fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
# [left, right, width, height]
axes1.plot(x,y)
plt.savefig("basicPlot.png")
```
